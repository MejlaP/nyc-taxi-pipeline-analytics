```python
from pyspark.sql.functions import *
```

```python
df = spark.read.table("nyctaxi.02_silver.yellow_trips_enriched")
```

```python
# Analyzes which taxi vendor generates the highest total revenue
df.groupBy("vendor").\
    agg(round(sum("total_amount"),2).alias("total_revenue")).\
    orderBy("total_revenue", ascending=False).\
    limit(3).\
    display()
```

```python
# Analyzes which taxi vendor generates the lowest total revenue
df.groupBy("vendor").\
    agg(round(sum("total_amount"),2).alias("total_revenue")).\
    orderBy("total_revenue").\
    limit(1).\
    display()
```

```python
# Identifies the most popular pickup boroughs by trip count
df.groupBy("pu_borough").\
    agg(count("*").alias("count_of_trips")).\
    orderBy("count_of_trips", ascending=False).\
    display()
```

```python
# Identifies the most common journeys
df.groupBy(concat("pu_borough", lit("-->"), "do_borough").alias("journey")).\
    agg(count("*").alias("count_of_trips")).\
    orderBy("count_of_trips", ascending=False).\
    display()
```
