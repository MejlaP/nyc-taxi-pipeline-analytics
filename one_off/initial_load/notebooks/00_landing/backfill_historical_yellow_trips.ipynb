```python
import os # operating system
```

```python
# NOTE: This script performs validation of the **historical NYC Yellow Taxi trip data** in the landing zone. It ensures data availability before downstream processing.
```

```python
# Define the list of months to be processed
dates_to_process = ['2025-06', '2025-07', '2025-08', '2025-09', '2025-10', '2025-11']

# Define the base directory path where the data files are stored
base_path = "/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow"

# Iterate through each month to verify the presence of the corresponding Parquet file
for date in dates_to_process:
    # Construct the full path to the Parquet file for the current month
    file_path = f"{base_path}/{date}/yellow_tripdata_{date}.parquet"
    
    # Check if the file exists at the specified location
    # If the file is missing, we raise an error, causing the notebook/task to fail immediately
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Source file not found for {date} at {file_path}")
            
    print(f"Validated: {file_path}")

# If the loop finishes without errors, it means all files are present and ready for processing
print("All files validated. Pipeline is ready to proceed to the transformation phase.")
```
