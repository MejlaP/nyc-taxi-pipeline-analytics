```python
from pyspark.sql.functions import current_timestamp
```

```python
# Load all raw parquet files from the landing zone using a wildcard (*)
df = spark.read.format("parquet").load("/Volumes/nyctaxi/00_landing/data_sources/nyctaxi_yellow/*")
```

```python
# Add a metadata column with the current timestamp to track when the data was processed
df = df.withColumn("processed_timestamp", current_timestamp())
```

```python
# Write the data into the Bronze layer as a permanent Delta table
# Note: 'overwrite' mode is used here to safely recreate the table during development
df.write.mode("overwrite").saveAsTable("nyctaxi.01_bronze.yellow_trips_raw")
```
